```python
from IPython.display import display, Markdown, clear_output
from openai import OpenAI
base_url = "https://certain-inc-newest-might.trycloudflare.com"

client = OpenAI(
    base_url=f"{base_url}/v1",
    api_key="dummy"
)

stream = client.chat.completions.create(
    model="qwen3.5:9b",
    messages=[
        {"role":"user","content":"Respond with a sonnet about the beauty of nature."}
    ],
    stream=True
)

output = []
handle = display(Markdown(""), display_id=True)
for chunk in stream:
    content = chunk.choices[0].delta.content
    if content:
        output.append(content)
        handle.update(Markdown("".join(output)))
```
