```python
from IPython.display import display, Markdown, clear_output
from openai import OpenAI
base_url = "https://certain-inc-newest-might.trycloudflare.com"

client = OpenAI(
    base_url=f"{base_url}/v1",
    api_key="dummy"
)

stream = client.chat.completions.create(
    model="qwen3.5:9b",
    messages=[
        {"role":"user","content":"Write a 1000 line story about a magical kingdom where animals can talk and humans "
        "are the minority. The story should be engaging, with a clear plot, interesting characters, and a satisfying "
        "conclusion. Include dialogue, descriptions of the setting, and any magical elements that make the kingdom unique."}
    ],
    stream=True
)

output = []
handle = display(Markdown(""), display_id=True)
for chunk in stream:
    content = chunk.choices[0].delta.content
    if content:
        output.append(content)
        handle.update(Markdown("".join(output)))
```
